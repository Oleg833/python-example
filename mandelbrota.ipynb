```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = 0
    n = 0
    while abs(z) <= 2 and n < max_iter:
        z = z**2 + c
        n += 1
    if n == max_iter:
        return 0
    else:
        return n

def create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter):
    image = np.zeros((height, width))

    for x in range(width):
        for y in range(height):
            real = x_min + (x / width) * (x_max - x_min)
            imag = y_min + (y / height) * (y_max - y_min)
            c = complex(real, imag)
            color = mandelbrot(c, max_iter)
            image[y, x] = color

    return image

def display_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter):
    mandelbrot_image = create_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter)

    plt.imshow(mandelbrot_image, cmap='viridis', extent=(x_min, x_max, y_min, y_max))
    plt.colorbar()
    plt.title('Mandelbrot Set')
    plt.show()

# Задайте розміри та параметри області для відображення
width, height = 800, 800
x_min, x_max = -2, 2
y_min, y_max = -2, 2
max_iter = 1000

# Відобразіть Множину Мандельброта
display_mandelbrot(width, height, x_min, x_max, y_min, y_max, max_iter)
```
