```python
def fibonacci_iterative(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for i in range(2, n+1):
        print(f"i={i-1}, n={n}, a={a},b={b}")
        a, b = b, a + b        
        print(f"return fibonacci {i-1}, a={a}, b=a+b={b}")
    return b

#  
print(fibonacci_iterative(10))  # : 55
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)  # Unbounded cache
def fibonacci(n):
    print("  fibonacci  n = ", n)
    if n <= 1: #  
        print(f" , n = {n},  {n}")
        return n
    else:
        result = fibonacci(n-1) + fibonacci(n-2) #  
        print(f"   n = {n}, {fibonacci(n-1)} + {fibonacci(n-2)} fibonacci({n}) = {result}")
        return result

print(fibonacci(10)) #  55
    
```

```python
from functools import lru_cache

# @lru_cache(maxsize=None)  # Unbounded cache
def fibonacci_memo(n, memo={}):
    print("  fibonacci  n = ", n)
    if n in memo:
        result = memo[n]
        print(f"Memorise fibonacci({n}) = {result}")
        return result
    if n <= 1:
        memo[n] = n
        print(f" , n = {n},  {n}")
        return n
    else:
        value = fibonacci_memo(n-1, memo) + fibonacci_memo(n-2, memo)
        memo[n] = value
        print(f"   n = {n}, {fibonacci_memo(n-1, memo)} + {fibonacci_memo(n-2, memo)} fibonacci({n}) = {value}")
        return value

print(fibonacci_memo(7))
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_triangle(vertices, ax):
    triangle = patches.Polygon(vertices, fill=False, edgecolor='black')
    ax.add_patch(triangle)

def midpoint(point1, point2):
    return [(point1[0] + point2[0]) / 2, (point1[1] + point2[1]) / 2]

def sierpinski(vertices, level, ax):
    draw_triangle(vertices, ax)
    if level > 0:
        sierpinski([vertices[0], midpoint(vertices[0], vertices[1]), midpoint(vertices[0], vertices[2])], level-1, ax)
        sierpinski([vertices[1], midpoint(vertices[0], vertices[1]), midpoint(vertices[1], vertices[2])], level-1, ax)
        sierpinski([vertices[2], midpoint(vertices[2], vertices[1]), midpoint(vertices[0], vertices[2])], level-1, ax)

def main():
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_axis_off()
    vertices = [[0, 0], [0.5, 0.75], [1, 0]]
    sierpinski(vertices, 3, ax)
    plt.show()

main()
```

```python
import turtle

def koch_curve(t, order, size):
    if order == 0:
        t.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
            koch_curve(t, order - 1, size / 3)
            t.left(angle)

def draw_koch_curve(order, size=300):
    window = turtle.Screen()
    window.bgcolor("white")

    t = turtle.Turtle()
    t.speed(0)  
    t.penup()
    t.goto(-size / 2, 0)
    t.pendown()

    koch_curve(t, order, size)

    window.mainloop()

#  
draw_koch_curve(3)
```

```python
import turtle

def koch_snowflake(turtle, order, size):
    if order == 0:
        turtle.forward(size)
    else:
        for angle in [60, -120, 60, 0]:
            koch_snowflake(turtle, order-1, size/3)
            turtle.left(angle)

#        
my_turtle = turtle.Turtle()
my_turtle.speed(3)
koch_snowflake(my_turtle, 3, 300)

turtle.done()
```

```python
def bubble_sort(lst):
    n = len(lst)
    for i in range(n-1):
        for j in range(0, n-i-1): 
            if lst[j] > lst[j+1] :
                lst[j], lst[j+1] = lst[j+1], lst[j] 
    return lst

numbers = [5, 3, 8, 4, 2]
bubble_sort(numbers)
```
